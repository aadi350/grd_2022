```python
import pandas as pd

pd.set_option('display.max_rows', 500)
```

```python
DATA_PATH = '../data/Country C.xlsx'

# read into Excel file (all sheets) as a dictionary of name:df
sheet_to_df_map = pd.read_excel(DATA_PATH, sheet_name=None)

sheet_to_df_map
```

```python
# don't load 'Source' sheet
# get list of "columns" for each sheet
# data is for 1996-2021, need to do each individually
for sheet in sheet_to_df_map.keys():
    print(f'{sheet}: {sheet_to_df_map[sheet].columns}')
```

```python
# 1996
sheet_to_df_map[' RR MAX & MIN 1996'].isna().sum()
```

```python
# drop all rows where index is null as these are aggregate in Excel
sheet_to_df_map[' RR MAX & MIN 1996'] = sheet_to_df_map[' RR MAX & MIN 1996'].dropna(axis=0, subset=[1996])
```
